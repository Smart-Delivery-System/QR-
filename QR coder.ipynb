```python
import tkinter as tk
from tkinter import *
from PIL import ImageTk, Image

r=tk.Tk()
r.title("Seamless and Intelligent Shipping")
r.geometry("1050x700")

Label = Label(r, text = 'SEAMLESS AND INTELLIGENT SHIPPING', font = ('Franklin Gothic Medium',30))

Label.pack()
Label.place(x=200,y=80)

load=Image.open('del1.png')
render = ImageTk.PhotoImage(load)
imgp = load.resize((300,300),Image.ANTIALIAS)
imag = ImageTk.PhotoImage(imgp)
img = tk.Label(image=imag)
img.image = render
img.place(x=600,y=150)

load1=Image.open('custum.png')
render1 = ImageTk.PhotoImage(load1)
imgp1 = load1.resize((300,300),Image.ANTIALIAS)
imag1 = ImageTk.PhotoImage(imgp1)
img1 = tk.Label(image=imag1)
img1.image = render1
img1.place(x=200,y=150)

def click1():
    M=Tk()
    M.geometry("500x500")
    load2=Image.open('del.jpg')
    render2 = ImageTk.PhotoImage(load2)
    imgp2 = load2.resize((300,300),Image.ANTIALIAS)
    imag2 = ImageTk.PhotoImage(imgp2)
    img2 = tk.Label(image=imag2)
    img2.image = render2
    img2.place(x=200,y=150)

    

def click():
    
    T=Tk()
    T.geometry("500x500")
    tk.Label(T, text='Enter the OTP',font=('Bahnschrift',10),width=20,height=20).grid(row=0)
    

    e1 = tk.Entry(T)
    

    e1.grid(row=0, column=1)
    
    tk.Button(T, text='Quit',width=20,height=2,command=T.destroy).grid(row=3,column=1, sticky=tk.W,pady=4)
    tk.Button(T, text='CONTINUE',width=20,height=2,command=click1).grid(row=3,column=5)

    
    
    
    
    
    
button5=Button(r,text='CUSTOMER', width=30,height=2,command=click)
button5.pack()
button5.place(x=250,y=500)
button1=Button(r,text='DELIVERY-MAN',width=30,height=2, command=click)
button1.place(x=650,y=500)
quitbutton=Button(text='QUIT',width=30,height=2,command=r.destroy)
quitbutton.place(x=460,y=600)


r.mainloop()


```
